```python
#carga desde un archivo .xlsx sin índice
df=pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

```python
#verificamos info de un df
df.info()
```

```python
#FILTRO UNO
#dentro de este filtro buscamos sacar de la columna NOMBRE_VENDEDOR, todas las respuestas que sean igual dentro de esa misma columna con LETICIA RAMIREZ HERNANDEZ
#estamos haciendo un filtro donde buscamos filas dentro de una columna
filtro1= df[df['NOMBRE_VENDEDOR']=='LETICIA RAMIREZ HERNANDEZ']
filtro1
```

```python
filtro1.to_csv('FILTRO UNO.csv')

from google.colab import files
files.download('FILTRO UNO.csv')
```

```python
#FILTRO DOS
#lo que se busca es usar la localización de las FILAS para imprimir valores
#queremos la fila 3,4, y 5, los datos de sas filas y de todas las columnas
filtro2 = df.iloc[3:5,:]                   #filas consecutivas
filtro2
```

```python
filtro2.to_csv('FILTRO DOS.csv')

from google.colab import files
files.download('FILTRO DOS.csv')
```

```python
#FILTRO TRES
#ahora, queremos busacr por medio de la ubicación de ínidces (localización) las columnas 1,3 y 5, y queremos todas sus filas
filtro3= df.iloc[:,[1,3,5]]
filtro3
```

```python
filtro3.to_csv('FILTRO TRES.csv')

from google.colab import files
files.download('FILTRO TRES.csv')
```

```python
#FILTRO CUATRO
#lo q primero que hacemos es asignar un índice con la finalidad de buscar dentro de esa columna que asiganmos, ciertas filas
#nombramos a la columna de NOMBRE_CLIENTE como nuestro índice, y dentro de ella buscamos el PÚBLICO EN GENERAL y OMAR RODRIGUEZ GOMEZ	 junto con la columna de PRECIO_UNITARIO
df1=pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=3)
filtro4= df1.loc[['PÚBLICO EN GENERAL', 'OMAR RODRIGUEZ GOMEZ'], ['PRECIO_UNITARIO']]
filtro4
```

```python
filtro4.to_csv('FILTRO CUATRO.csv')

from google.colab import files
files.download('FILTRO CUATRO.csv')
```

```python
#FILTRO CINCO
#lo que hacemos ahora es ver únicamente los primeros 4 registros de todo el df
filtro5 = df.head(4)
filtro5
```

```python
filtro5.to_csv('FILTRO CINCO.csv')

from google.colab import files
files.download('FILTRO CINCO.csv')
```

```python
#FILTRO 6
# ahora buscamos dentro de la columna de precio unitario, las filas que sean mayores a 1000
filtro6= df[df['PRECIO_UNITARIO']>1000]
filtro6
```

```python
filtro6.to_csv('FILTRO SEIS.csv')

from google.colab import files
files.download('FILTRO SEIS.csv')
```

```python
#FILTRO 7
#ahora buscamos dos "condicionales"
#queremos imprimir las filas que donde su precio unitario sea menor a 500 y que su fecha sea igual a 2022-01-05
#solo se imprimen las filas que cumplen con ambas condiciones
filtro7 = df[(df["PRECIO_UNITARIO"] < 550) & (df["FECHA_DOC"] == '2022-01-05')]
filtro7
```

```python
filtro7.to_csv('FILTRO SIETE.csv')

from google.colab import files
files.download('FILTRO SIETE.csv')
```

```python
#FILTRO 8
#lo que buscamos ahora es imprimir con dos conectores lógicos
#queremos buscar dentro del df los valores de COSTO_TOTAL_CALCULADO que sean menores a 500 o que su fecha sea igual a 2022-01-05
filtro8 =df[(df['COSTO_TOTAL_CALCULADO']<500) | (df['FECHA_DOC']=='2022-01-05')]
filtro8
```

```python
filtro8.to_csv('FILTRO OCHO.csv')

from google.colab import files
files.download('FILTRO OCHO.csv')
```

```python
#FILTRO 9
#buscamos buscar registros que no se cumplan con las condiciones
#buscamos todos los registros que su MARGEN_UNITARIO_CALCULADO NO SEA menor o igual a 670 y que su fecha NO SEA 2022-01-05
filtro9= df[~(df['MARGEN_UNITARIO_CALCULADO']<=670)&~(df['FECHA_DOC']=='2022-01-05')]
filtro9
```

```python
filtro9.to_csv('FILTRO NUEVE.csv')

from google.colab import files
files.download('FILTRO NUEVE.csv')
```
