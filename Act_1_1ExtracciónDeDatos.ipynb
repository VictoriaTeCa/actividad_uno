```python
#carga desde un archivo .xlsx sin índice
df=pd.read_excel('Datos de Facturación.xlsx')
df
```

```python
#verificamos info de un df
df.info()
```

```python
#CVE_CLPV filtrar los valores 1000 a 2000
df.CVE_CLPV = df.CVE_CLPV.replace({'O':'0', '6781|': '6781', '3612C':'3612', '      ': '0', '2385A': '2385', '2510A':'2510', '8440Q':'8440', 'M':'0'}, regex=False)

df['CVE_CLPV']=df['CVE_CLPV'].astype(float)
```

```python
filtro1_1 = df[(df["CVE_CLPV"] >= 1000) & (df["CVE_CLPV"] <= 2000)]
filtro1_1
```

```python
filtro1_1.to_csv('filtro 1_1 actividad_uno.csv')

from google.colab import files
files.download('filtro 1_1 actividad_uno.csv')
```

```python
# CVE_VEND  todas las claves exepto 5 y 4
filtro2_1 = df[~(df['CVE_VEND']==5) & ~(df['CVE_VEND']==4)]
filtro2_1
```

```python
filtro2_1.to_csv('filtro 2_1 actividad_uno.csv')

from google.colab import files
files.download('filtro 2_1 actividad_uno.csv')
```

```python
#corroboramos
corroborar = filtro2_1.groupby(['CVE_VEND'])['CVE_VEND'].count()
corroborar
```
